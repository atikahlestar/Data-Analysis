## Soal 1
Hitung tingkat retensi pelanggan untuk setiap level keanggotaan (Tier). Berdasarkan hasil tersebut, analisis dan jelaskan pola retensi yang Anda temukan. Apakah ada level keanggotaan yang menunjukkan tingkat retensi yang lebih tinggi atau lebih rendah dibandingkan yang lain? Jelaskan insight yang dapat diambil dari pola tersebut.

### Jawab :

```python
## Mengelompokkan berdasarkan Tier
retention_rate = df.groupby("Tier")["Retained"].mean() * 100
print(retention_rate)
```

```python
## Visualisasi
plt.figure(figsize=(10, 6))
bars = plt.bar(retention_rate.index, retention_rate.values)
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Posisi x (tengah batang)
        height,                             # Posisi y (di atas batang)
        f'{height:.2f}%',                   # Format angka (2 desimal dan tanda %)
        ha='center', va='bottom', fontsize=10  # Penyesuaian tampilan
    )

plt.xlabel("Tier")
plt.ylabel("Tingkat Retensi (%)")
plt.title("Tingkat Retensi Pelanggan Berdasarkan Tier")
plt.ylim(0, 120)  # Menambah ruang di atas untuk teks
plt.show()
```

### Analisis :

- Pelanggan dengan tier keanggotaan lebih tinggi (Gold dan Platinum) memiliki tingkat retensi yang lebih tinggi dibandingkan dengan tier lebih rendah (Blue dan Silver).

- Ini mungkin mencerminkan bahwa pelanggan dengan akses ke lebih banyak manfaat atau layanan premium merasa lebih puas dan cenderung tetap bertahan.


## Soal 2
Berapa rata-rata pengeluaran bulanan pelanggan untuk setiap level keanggotaan (Tier)? Apakah pelanggan dengan tier yang lebih tinggi cenderung memiliki pengeluaran bulanan yang lebih tinggi? Jelaskan apa yang dapat dipelajari dari perbedaan pengeluaran ini.

### Jawab :

```python
print(df["Monthly_Spend_IDR"].head())
```

```python
df["Monthly_Spend_IDR"] = df["Monthly_Spend_IDR"].astype(str)
```

```python
# Bersihkan data dengan hati-hati
df["Monthly_Spend_IDR"] = (
    df["Monthly_Spend_IDR"]
    .str.replace(r"[^\d]", "", regex=True)  # Hapus semua karakter non-numerik
    .astype(float)  # Konversi ke float
)
```

```python
## Menghitung rata-rata pengeluaran
pd.options.display.float_format = '{:,.2f}'.format  # Format dengan 2 desimal dan koma pemisah ribuan
average_spend = df.groupby("Tier")["Monthly_Spend_IDR"].mean()
print(average_spend)
```

### Analisis :

Pelanggan dengan tingkat keanggotaan yang lebih tinggi (Gold dan Platinum) secara signifikan memiliki pengeluaran bulanan yang lebih besar dibandingkan dengan pelanggan di tingkatan yang lebih rendah (Blue dan Silver). Dari sini, dapat disimpulkan bahwa pelanggan di tingkatan yang lebih tinggi cenderung berkontribusi lebih banyak secara finansial.

Analisis ini menunjukkan bahwa program khusus seperti diskon atau fasilitas tambahan yang ditawarkan kepada pelanggan Gold dan Platinum mungkin berhasil mendorong mereka untuk melakukan lebih banyak pengeluaran. Strategi yang menargetkan peningkatan pengeluaran di kalangan pelanggan tingkat atas dapat terus dikembangkan. Di sisi lain, mungkin ada peluang untuk meningkatkan pengeluaran di tingkat keanggotaan yang lebih rendah dengan penawaran insentif atau promosi yang lebih terarah.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Age", y="Retained", hue="Tier", marker="o")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Retention Rate", fontsize=12)
plt.title("Customer Retention Rate by Age and Tier", fontsize=14)
plt.legend(title="Tier")
plt.grid(True)
plt.show()
```

Karena saya tidak melihat pola yang jelas, saya mencoba untuk membuat kategori umur. Dimana pengelompokan tersebut memberikan hasil seperti ini


```python
# Membuat Kelompok Usia (Age Groups)
bins = [18, 24, 34, 44, 54, 64] # Batas Usia
Labels = ["18-24", "25-34", "35-44", "45-54", "55-64"] # Label Kelompok
df["Age_Group"] = pd.cut(df["Age"], bins=bins, labels=Labels, right=True)

print(df)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Age_Group", y="Retained", hue="Tier", marker="o")
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Retention Rate", fontsize=12)
plt.title("Customer Retention Rate by Age Group and Tier")
plt.legend(title="Tier")
```

### Analisis :

1. Pelanggan Platinum dan Gold adalah kelompok yang paling loyal, dan strategi untuk mempertahankan pelanggan ini harus diprioritaskan.
2. Kelompok usia 25-34 memerlukan perhatian khusus karena tingkat retensi mereka lebih rendah dibandingkan kelompok lainnya.
3. Pelanggan Blue memiliki tingkat retensi terendah, sehingga perlu ditingkatkan dengan strategi pemasaran yang efektif atau insentif tambahan untuk meningkatkan loyalitas.

### Analisis

Berikut adalah analisis dari grafik yang menunjukkan hubungan antara lama waktu berlangganan (Tenure) dan tingkat retensi (Retention Rate) berdasarkan tingkatan keanggotaan (Tier):

1. Pola Umum:
  - Grafik memperlihatkan bahwa tingkat retensi memiliki variasi di setiap kelompok Tier seiring bertambahnya waktu (Tenure).
  - Untuk beberapa kelompok, tingkat retensi meningkat saat Tenure bertambah, sementara pada kelompok lainnya terdapat fluktuasi atau penurunan.
2. Performa Berdasarkan Tier:
  - Platinum Tier:
    - Secara konsisten menunjukkan tingkat retensi yang lebih tinggi dibandingkan Tier lainnya.
    - Memiliki tren peningkatan yang signifikan dalam periode awal (1-6 bulan), kemudian cenderung stabil meskipun terdapat fluktuasi kecil.
  - Gold Tier:
    - Cenderung berada di urutan kedua setelah Platinum dalam hal tingkat retensi.
    - Tingkat retensi meningkat dalam periode awal dan menunjukkan fluktuasi moderat setelahnya.
  - Silver Tier:
    - Tingkat retensi menurun secara bertahap dari awal hingga pertengahan periode (3-8 bulan), namun terdapat sedikit peningkatan menjelang akhir.
  - Blue Tier:
    - Memiliki tingkat retensi paling rendah di antara semua Tier.
    - Tingkat retensi tetap rendah secara konsisten tanpa peningkatan yang signifikan.
3. Korelasi Tenure dan Retensi:
  - Pada Platinum dan Gold Tier, terdapat hubungan positif yang lebih jelas antara Tenure yang lebih lama dan tingkat retensi, terutama pada bulan-bulan awal.
  - Sebaliknya, pada Silver dan Blue Tier, korelasi tersebut tidak terlalu terlihat. Pelanggan cenderung tidak menunjukkan peningkatan retensi meskipun waktu berlangganan bertambah.
4. Kesimpulan:
  - Tingkatan keanggotaan memiliki dampak signifikan terhadap retensi pelanggan. Pelanggan di Platinum dan Gold Tier lebih cenderung bertahan dibandingkan yang berada di Blue dan Silver Tier.
  - Strategi peningkatan retensi perlu difokuskan pada pelanggan Blue dan Silver, terutama dalam periode awal berlangganan (1-6 bulan) untuk mencegah penurunan retensi yang lebih besar.

Penggunaan skala log pada sumbu y membantu memperjelas rentang pengeluaran yang bervariasi antar tier.

## Soal 10

Apakah ada tier tertentu yang mengalami penurunan atau peningkatan tingkat retensi ketika usia pelanggan meningkat? Bagaimana pengaruh usia terhadap tingkat retensi pada tiap tier?

SignalID tertarik untuk memahami apakah ada penurunan retensi di kalangan pelanggan yang lebih tua atau lebih muda di setiap tingkatan keanggotaan. Ini akan membantu mereka menyesuaikan program loyalitas atau fasilitas yang lebih sesuai dengan usia pelanggan.


### Jawab

```python
## Menghitung tingkat retensi berdasarkan usia (Age) untuk setiap Tier
age_retention = df.groupby(["Age", "Tier"])["Retained"].mean().reset_index()
print(age_retention)
```

```python
# Membuat subplots untuk setiap Tier
tiers = ["Blue", "Silver", "Gold", "Platinum"]
fig, axes = plt.subplots(2, 2, figsize=(10, 6), sharey=True)
axes = axes.flatten()

for i, tier in enumerate(tiers):
    tier_data = age_retention[age_retention["Tier"] == tier]
    sns.lineplot(
        data=tier_data,
        x="Age",
        y="Retained",
        marker='o',
        ax=axes[i],
        label=f"{tier} Tier"
    )
    axes[i].set_title(f"Retention Rate by Age for {tier} Tier")
    axes[i].set_xlabel("Age")
    axes[i].set_ylabel("Retention Rate")
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

### Analisis

- Di tier Silver, pelanggan yang lebih tua menunjukkan retensi yang lebih tinggi, sehingga mungkin program loyalitas yang lebih sesuai untuk pelanggan senior akan bermanfaat.
- Di tier Gold, pelanggan muda lebih loyal, sementara pelanggan yang lebih tua cenderung mengalami penurunan retensi. Mungkin perlu dilakukan strategi khusus untuk meningkatkan kepuasan pelanggan yang lebih tua di tingkatan ini.
